```python
import datetime as dt
import pandas as pd
```

```python
filename_news = "../Signals/Combined_News_DJIA_extended.csv"
filename_label = "../Signals/Case B - interval - extended/Plus4/DJIA_signal_plus4B_extended.xlsx"
final_output = "../Signals/Case B - interval - extended/Plus4/Combined_News_DJIA_plus4B_extended.csv"
```

```python
news = pd.read_csv(filename_news)
news = news.drop('Label', axis=1)
news['Date']=news['Date'].astype(str)
news
```

```python
label = pd.read_excel(filename_label)
label['Date']=label['Date'].astype(str)
label
```

```python
combinedNewsStock = pd.merge(label,news,on='Date')
combinedNewsStock = combinedNewsStock[['Date','Label','Top1','Top2','Top3','Top4','Top5','Top6','Top7','Top8','Top9','Top10','Top11','Top12','Top13','Top14','Top15','Top16','Top17','Top18','Top19','Top20','Top21','Top22','Top23','Top24','Top25']]
combinedNewsStock = combinedNewsStock.sort_values(by='Date')
```

```python
combinedNewsStock = combinedNewsStock.drop_duplicates(subset=['Date'], keep='first')
combinedNewsStock.shape
```

```python
combinedNewsStock.to_csv(final_output, index=False)
```
